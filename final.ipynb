```python
hp = EPL_data.groupby(['Season','HomeTeam'])['FTR'].count().rename("HP")
hw = EPL_data[EPL_data['FTR'] =='H'].groupby(['Season','HomeTeam'])['FTR'].count().rename("HW")
hl = EPL_data[EPL_data['FTR'] =='A'].groupby(['Season','HomeTeam'])['FTR'].count().rename("HL")
hd = EPL_data[EPL_data['FTR'] =='D'].groupby(['Season','HomeTeam'])['FTR'].count().rename("HD")
home_games=pd.concat([hp,hw,hl,hd],axis=1).reset_index()
home_games.fillna(0,inplace=True)
home_games[['HP','HL','HD','HW']]=home_games[['HP','HL','HD','HW']].astype(int)

home_score = EPL_data.groupby(['Season', 'HomeTeam'])[['FTAG', 'FTHG']].sum().astype(int).reset_index()

home_score.rename(columns={'FTHG':'HG For','FTAG':'HG Against'},inplace=True)

home_df=pd.merge(left=home_games,right=home_score,on=['Season','HomeTeam'],how='left')
home_df
```

```python
ap = EPL_data.groupby(['Season','AwayTeam'])['FTR'].count().rename("AP")
aw = EPL_data[EPL_data['FTR'] =='A'].groupby(['Season','AwayTeam'])['FTR'].count().rename("AW")
al = EPL_data[EPL_data['FTR'] =='H'].groupby(['Season','AwayTeam'])['FTR'].count().rename("AL")
ad = EPL_data[EPL_data['FTR'] =='D'].groupby(['Season','AwayTeam'])['FTR'].count().rename("AD")
away_games=pd.concat([ap,aw,al,ad],axis=1).reset_index()
away_games.fillna(0,inplace=True)   ##NaN comes where theere are 0 games as groupby misses that entry
away_games[['AP','AL','AD','AW']]=away_games[['AP','AL','AD','AW']].astype(int)

away_score=EPL_data.groupby(['Season','AwayTeam'])[['FTAG','FTHG']].sum().astype(int).reset_index()
away_score.rename(columns={'FTHG':'AG Against','FTAG':'AG For'},inplace=True)

away_df=pd.merge(left=away_games,right=away_score,on=['Season','AwayTeam'],how='left')
away_df
```

```python
pt_table = pd.merge(left=home_df,right=away_df,left_on=['Season','HomeTeam'],right_on=['Season','AwayTeam']).drop(['AwayTeam'],axis=1)
pt_table.rename(columns={'HomeTeam':'Team'},inplace=True)
pt_table['Total Played'] = pt_table['HP'] + pt_table['AP']
pt_table['Total Wins'] = pt_table['HW'] + pt_table['AW']
pt_table['Total Loss'] = pt_table['HL'] + pt_table['AL']
pt_table['Total Draw'] = pt_table['HD'] + pt_table['AD']
pt_table['GF'] = pt_table['HG For'] + pt_table['AG For']
pt_table['GA'] = pt_table['HG Against'] + pt_table['AG Against']
pt_table['GD'] = pt_table['GF'] - pt_table['GA'] #goal difference
pt_table['Total_Points'] = 3*pt_table['Total Wins'] + pt_table['Total Draw']  # 3 Points for a Win and 1 Point for a Draw.
```

```python
## Ranking the teams
pt_table['Rank']= pt_table.groupby('Season')['Total_Points'].rank(method='min',ascending=False)
pt_table['fin'] = pt_table.groupby(['Season','Rank'])['GD'].rank(method='first',ascending=False)  # If points are same, use goal difference
pt_table['Rank']=np.where((pt_table['fin'] > 1),(pt_table['Rank']+pt_table['fin'] - 1),pt_table['Rank'])
pt_table.drop('fin',axis=1,inplace=True)
```

```python
import pandas as pd
import plotly.express as px
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

EPL_data = pd.read_csv('./english-premier-league-results/results.csv', encoding='unicode_escape')

EPL_data.dropna(inplace=True)

# Define the Dash app
app = dash.Dash(__name__)

#Define the app layout
app.layout = html.Div([
    html.H1("English Premier League Dashboard"),
    html.Div([
        html.Label("Select Season:"),
        dcc.Dropdown(
            id='season-dropdown',
            options=[{'label': season, 'value': season} for season in EPL_data['Season'].unique()],
            value='1993-94',  # Default value
            multi=False
        ),
    ]),
    
    html.Div([
        html.Label("Select Team(s):"),
        dcc.Dropdown(
            id='team-dropdown',
            options=[{'label': team, 'value': team} for team in sorted(pd.concat([EPL_data['HomeTeam'], EPL_data['AwayTeam']]).unique())],
            value=['Arsenal'],  # Default value
            multi=True
        ),
    ]),
    
   html.Div([
        html.Label("Select Season Range:"),
        dcc.RangeSlider(
            id='season-range-slider',
            min=EPL_data.index.min(),
            max=EPL_data.index.max(),
            step=1,
            value=[EPL_data.index.min(), EPL_data.index.max()],
            marks={i: f'Season {season}' for i, season in enumerate(EPL_data['Season'].unique())}
        )
    ]),
      html.Div([
        dcc.Graph(id='goals-per-season-graph')
    ]),
    html.Div([
        html.Label("Ask a question about the dataset:"),
        dcc.Input(
            id='question-input',
            type='text',
            value='',
            placeholder='Type your question here...'
        ),
        html.Button('Submit', id='submit-button', n_clicks=0),
        html.Div(id='question-output')
    ])
])

# Define callback to update goals per season graph
# Define callback to update goals per season graph
@app.callback(
    Output('goals-per-season-graph', 'figure'),
    [Input('season-dropdown', 'value'),
     Input('team-dropdown', 'value'),
     Input('season-range-slider', 'value')]
)
def update_goals_per_season_graph(selected_season, selected_teams, season_range):
    try:
        filtered_data = EPL_data[(EPL_data['Season'] == selected_season) &
                                 (EPL_data['HomeTeam'].isin(selected_teams) | EPL_data['AwayTeam'].isin(selected_teams)) &
                                 (EPL_data.index >= season_range[0]) & (EPL_data.index <= season_range[1])]
        goals_per_season = filtered_data.groupby('Season')[['FTHG', 'FTAG']].sum().reset_index()
        goals_per_season['Total_Goals'] = goals_per_season['FTHG'] + goals_per_season['FTAG']
        fig = px.bar(goals_per_season, x='Season', y='Total_Goals', title='Goals Scored per Season')
        return fig
    except Exception as e:
        print(e)  # Print the error for debugging purposes
        return px.bar(title='Error: Unable to generate graph')  # Return a placeholder graph in case of error

# Define callback to answer questions about the dataset
@app.callback(
    Output('question-output', 'children'),
    [Input('submit-button', 'n_clicks')],
    [dash.dependencies.State('question-input', 'value')]
)
def answer_question(n_clicks, question):
    if n_clicks > 0:
        if question.strip() == '':
            return 'Please type a question.'
        elif 'referee' in question.lower() and 'red cards' in question.lower():
            referee_red_cards = EPL_data['Referee'].value_counts().idxmax()
            return f"The referee who gave the most red cards is: {referee_red_cards}"
        elif 'comebacks' in question.lower() or 'season' in question.lower():
            season_most_comebacks = EPL_data[EPL_data['HTR'] != EPL_data['FTR']]['Season'].value_counts().idxmax()
            return f"The season with the most comebacks is: {season_most_comebacks}"
        elif 'standings' in question.lower() or 'end of the season' in question.lower():
            # Perform calculations for standings at the end of the season
            # Replace this with your specific logic
            return "Standings at the end of the season: (Replace this with actual standings)"
        else:
            return "Sorry, I couldn't understand your question."

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)

```
